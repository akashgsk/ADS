```python
class Employee:
    def __init__(self, name, designation, salary):
        self.name = name
        self.designation = designation
        self.salary = salary
        self.next = None

class EmployeeLinkedList:
    def __init__(self):
        self.head = None

    def insert_employee(self, employee):
        new_node = Employee(employee['name'], employee['designation'], employee['salary'])
        new_node.next = self.head
        self.head = new_node

    def search_employee(self, employee_name):
        current = self.head
        while current:
            if current.name == employee_name:
                return current
            current = current.next
        return None

    def display_average_salary_manager(self):
        total_salary = 0
        manager_count = 0
        current = self.head

        while current:
            if current.designation.lower() == 'manager':
                total_salary += current.salary
                manager_count += 1
            current = current.next

        if manager_count > 0:
            average_salary = total_salary / manager_count
            print(f"Average salary of Managers: {average_salary}")
        else:
            print("No Managers found.")

# Example usage
employee_list = EmployeeLinkedList()

# Insert N employees into the linked list
employees_data = [
    {'name': 'John', 'designation': 'Manager', 'salary': 60000},
    {'name': 'Alice', 'designation': 'Developer', 'salary': 50000},
    # Add more employees as needed
]

for employee_data in employees_data:
    employee_list.insert_employee(employee_data)

# Search for an employee and display details
search_employee_name = 'John'
found_employee = employee_list.search_employee(search_employee_name)
if found_employee:
    print(f"Employee found - Name: {found_employee.name}, Designation: {found_employee.designation}, Salary: {found_employee.salary}")
else:
    print(f"Employee with name {search_employee_name} not found.")

# Display average salary of Managers
employee_list.display_average_salary_manager()
```

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.left = self.right = None

def insert(root, key):
    if root is None:
        return Node(key)
    else:
        if key < root.key:
            root.left = insert(root.left, key)
        else:
            root.right = insert(root.right, key)
    return root

def delete(root, key):
    if root is None:
        return root

    if key < root.key:
        root.left = delete(root.left, key)
    elif key > root.key:
        root.right = delete(root.right, key)
    else:
        # Node with one child or no child
        if root.left is None:
            return root.right
        elif root.right is None:
            return root.left

        # Node with two children, get the inorder successor
        root.key = minValueNode(root.right).key

        # Delete the inorder successor
        root.right = delete(root.right, root.key)

    return root

def minValueNode(node):
    current = node
    while current.left is not None:
        current = current.left
    return current

def level_order_traversal(root):
    if root is None:
        return

    queue = [root]
    while queue:
        current = queue.pop(0)
        print(current.key, end=" ")

        if current.left is not None:
            queue.append(current.left)
        if current.right is not None:
            queue.append(current.right)

# Example usage
root = None
keys = [50, 30, 70, 20, 40, 60, 80]

for key in keys:
    root = insert(root, key)

print("BST before deletion:")
level_order_traversal(root)

# Delete a node
key_to_delete = 30
root = delete(root, key_to_delete)
print(f"\nBST after deleting node with key {key_to_delete}:")
level_order_traversal(root)
```

```python
def preorder_traversal(root):
    if root is not None:
        print(root.key, end=" ")
        preorder_traversal(root.left)
        preorder_traversal(root.right)

def inorder_traversal(root):
    if root is not None:
        inorder_traversal(root.left)
        print(root.key, end=" ")
        inorder_traversal(root.right)

def postorder_traversal(root):
    if root is not None:
        postorder_traversal(root.left)
        postorder_traversal(root.right)
        print(root.key, end=" ")
```

```python
def dfs(graph, node, visited):
    if node not in visited:
        print(node, end=" ")
        visited.add(node)
        for neighbor in graph[node]:
            dfs(graph, neighbor, visited)

# Example usage:
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F','D','E'],
    'D': ['F'],
    'E': ['F'],
    'F': []
}

visited = set()
print("DFS on graph starting with vertex 'A' :")
dfs(graph, 'A', visited)
```

```python
def bfs(graph, start):
    visited = set()
    queue = [start]

    while queue:
        node = queue.pop(0)
        if node not in visited:
            print(node, end=" ")
            visited.add(node)
            queue.extend(neighbor for neighbor in graph[node] if neighbor not in visited)
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F','D','E'],
    'D': ['F'],
    'E': ['F'],
    'F': []
}
print("BFS on graph starting with vertex 'A' :")
bfs(graph, 'A')
```

```python
# 3. Write a program to insert N employee information into the Binary tree. Write a function to
# a. Search an Employee and display his details.
# b. Display the Average salary of Employees between the age 20 and 25.

class Employee:
    def __init__(self, name, age, salary):
        self.name = name
        self.age = age
        self.salary = salary
        self.left = self.right = None

def insert_employee(root, employee):
    if root is None:
        return Employee(employee['name'], employee['age'], employee['salary'])
    else:
        if employee['age'] < root.age:
            root.left = insert_employee(root.left, employee)
        else:
            root.right = insert_employee(root.right, employee)
    return root

def search_employee(root, employee_name):
    if root is None or root.name == employee_name:
        return root
    if root.name < employee_name:
        return search_employee(root.right, employee_name)
    return search_employee(root.left, employee_name)

def calculate_average_salary(root, age_range_start, age_range_end):
    total_salary = 0
    employee_count = 0

    def in_age_range(employee_age):
        return age_range_start <= employee_age <= age_range_end

    def traverse_and_calculate(node):
        nonlocal total_salary, employee_count
        if node:
            if in_age_range(node.age):
                total_salary += node.salary
                employee_count += 1
            traverse_and_calculate(node.left)
            traverse_and_calculate(node.right)

    traverse_and_calculate(root)

    if employee_count > 0:
        average_salary = total_salary / employee_count
        return average_salary
    else:
        return None

# Example usage
root = None

# Insert N employees into the binary tree
employees_data = [
    {'name': 'John', 'age': 25, 'salary': 60000},
    {'name': 'Alice', 'age': 22, 'salary': 50000},
    # Add more employees as needed
]

for employee_data in employees_data:
    root = insert_employee(root, employee_data)

# Search for an employee and display details
search_employee_name = 'John'
found_employee = search_employee(root, search_employee_name)
if found_employee:
    print(f"Employee found - Name: {found_employee.name}, Age: {found_employee.age}, Salary: {found_employee.salary}")
else:
    print(f"Employee with name {search_employee_name} not found.")

# Display average salary of employees between age 20 and 25
age_range_start = 20
age_range_end = 25
average_salary = calculate_average_salary(root, age_range_start, age_range_end)

if average_salary is not None:
    print(f"Average salary of employees between age {age_range_start} and {age_range_end}: {average_salary}")
else:
    print(f"No employees found in the specified age range.")
```

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = self.right = None

def count_leaf_nodes(node):
    if node is None:
        return 0
    if node.left is None and node.right is None:
        return 1
    return count_leaf_nodes(node.left) + count_leaf_nodes(node.right)

def count_total_nodes(node):
    if node is None:
        return 0
    return 1 + count_total_nodes(node.left) + count_total_nodes(node.right)

def display_all_values(node):
    if node is not None:
        print(node.value, end=" ")
        display_all_values(node.left)
        display_all_values(node.right)

# Example usage
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

# a. Number of leaf nodes
leaf_node_count = count_leaf_nodes(root)
print(f"a. Number of leaf nodes: {leaf_node_count}")

# b. Total number of nodes
total_node_count = count_total_nodes(root)
print(f"b. Total number of nodes: {total_node_count}")

# c. Display all values of the nodes
print("c. Values of the nodes:")
display_all_values(root)
```

```python
# 5. Write a program to insert N employee information into the Hash Table. Hash value should be the age of the employee.
# a. Write a function to Search an Employee and display his details.
# b. Display the Average salary of Employees between the age 22 and 25.

class Employee:
    def __init__(self, name, age, salary):
        self.name = name
        self.age = age
        self.salary = salary

class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        return key % self.size

    def insert_employee(self, employee):
        hash_value = self.hash_function(employee.age)
        if self.table[hash_value] is None:
            self.table[hash_value] = [employee]
        else:
            self.table[hash_value].append(employee)

    def search_employee(self, target_age):
        hash_value = self.hash_function(target_age)
        if self.table[hash_value] is not None:
            for employee in self.table[hash_value]:
                if employee.age == target_age:
                    return employee
        return None

    def calculate_average_salary(self, age_range_start, age_range_end):
        total_salary = 0
        employee_count = 0

        for age in range(age_range_start, age_range_end + 1):
            hash_value = self.hash_function(age)
            if self.table[hash_value] is not None:
                for employee in self.table[hash_value]:
                    if age_range_start <= employee.age <= age_range_end:
                        total_salary += employee.salary
                        employee_count += 1

        if employee_count > 0:
            average_salary = total_salary / employee_count
            return average_salary
        else:
            return None

# Example usage
hash_table_size = 10
employee_table = HashTable(hash_table_size)

# Insert N employees into the hash table
employees_data = [
    {'name': 'John', 'age': 25, 'salary': 60000},
    {'name': 'Alice', 'age': 22, 'salary': 50000},
    # Add more employees as needed
]

for employee_data in employees_data:
    employee = Employee(employee_data['name'], employee_data['age'], employee_data['salary'])
    employee_table.insert_employee(employee)

# Search for an employee and display details
search_age = 25
found_employee = employee_table.search_employee(search_age)
if found_employee:
    print(f"Employee found - Name: {found_employee.name}, Age: {found_employee.age}, Salary: {found_employee.salary}")
else:
    print(f"No employee found with age {search_age}.")

# Display average salary of employees between age 22 and 25
age_range_start = 22
age_range_end = 25
average_salary = employee_table.calculate_average_salary(age_range_start, age_range_end)

if average_salary is not None:
    print(f"Average salary of employees between age {age_range_start} and {age_range_end}: {average_salary}")
else:
    print(f"No employees found in the specified age range.")




```

```python
from queue import Queue

class StackUsingQueues:
    def __init__(self):
        self.queue1 = Queue()
        self.queue2 = Queue()

    def push(self, value):
        # Move all elements from queue1 to queue2
        while not self.queue1.empty():
            self.queue2.put(self.queue1.get())

        # Push the new element into queue1
        self.queue1.put(value)

        # Move all elements back from queue2 to queue1
        while not self.queue2.empty():
            self.queue1.put(self.queue2.get())

    def pop(self):
        # Check if the stack is empty
        if self.queue1.empty():
            print("Stack is empty. Cannot perform POP operation.")
            return None

        # Pop the top element from queue1
        popped_value = self.queue1.get()
        return popped_value

# Example usage
stack = StackUsingQueues()

# Push elements onto the stack
stack.push(1)
stack.push(2)
stack.push(3)

# Pop elements from the stack
print("Popped:", stack.pop())  # Output: Popped: 3
print("Popped:", stack.pop())  # Output: Popped: 2

# Push more elements
stack.push(4)
stack.push(5)

# Pop elements again
print("Popped:", stack.pop())  # Output: Popped: 5
print("Popped:", stack.pop())  # Output: Popped: 4
print("Popped:", stack.pop())  # Output: Popped: 1

# Attempt to pop from an empty stack
print("Popped:", stack.pop())  # Output: Stack is
```

```python
class QueueUsingStacks:
    def __init__(self):
        self.stack1 = []
        self.stack2 = []

    def push(self, value):
        # Push the new element onto stack1
        self.stack1.append(value)

    def pop(self):
        # If stack2 is empty, transfer elements from stack1 to stack2
        if not self.stack2:
            while self.stack1:
                self.stack2.append(self.stack1.pop())

        # If stack2 is still empty, the queue is empty
        if not self.stack2:
            print("Queue is empty. Cannot perform POP operation.")
            return None

        # Pop the front element from stack2
        popped_value = self.stack2.pop()
        return popped_value

# Example usage
queue = QueueUsingStacks()

# Push elements into the queue
queue.push(1)
queue.push(2)
queue.push(3)

# Pop elements from the queue
print("Popped:", queue.pop())  # Output: Popped: 1
print("Popped:", queue.pop())  # Output: Popped: 2

# Push more elements
queue.push(4)
queue.push(5)

# Pop elements again
print("Popped:", queue.pop())  # Output: Popped: 3
print("Popped:", queue.pop())  # Output: Popped: 4
print("Popped:", queue.pop())  # Output: Popped: 5

# Attempt to pop from an empty queue
print("Popped:", queue.pop())  # Output: Queue is empty. Cannot perform POP operation.
```

```python
# Write a program to insert N Student information into the Doubly Linked List. Write a function to
# a. Search a Student and display his details.
# b. Display the details of the students who have scored above 90 in Mathâ€™s and Science.

class Node:
    def __init__(self, roll_number, name, math_score, science_score):
        self.roll_number = roll_number
        self.name = name
        self.math_score = math_score
        self.science_score = science_score
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def insert_student(self, roll_number, name, math_score, science_score):
        new_node = Node(roll_number, name, math_score, science_score)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            new_node.prev = current

    def search_student(self, target_roll_number):
        current = self.head
        while current:
            if current.roll_number == target_roll_number:
                return current
            current = current.next
        return None

    def display_high_scorers(self):
        current = self.head
        print("Students who scored above 90 in Math and Science:")
        while current:
            if current.math_score > 90 and current.science_score > 90:
                print(f"Roll Number: {current.roll_number}, Name: {current.name}, Math Score: {current.math_score}, Science Score: {current.science_score}")
            current = current.next

# Example usage
students_list = DoublyLinkedList()

# Insert N students into the doubly linked list
students_data = [
    {'roll_number': 1, 'name': 'John', 'math_score': 95, 'science_score': 92},
    {'roll_number': 2, 'name': 'Alice', 'math_score': 88, 'science_score': 94},
    # Add more students as needed
]

for student_data in students_data:
    students_list.insert_student(student_data['roll_number'], student_data['name'], student_data['math_score'], student_data['science_score'])

# Search for a student and display details
search_roll_number = 1
found_student = students_list.search_student(search_roll_number)
if found_student:
    print(f"Student found - Roll Number: {found_student.roll_number}, Name: {found_student.name}, Math Score: {found_student.math_score}, Science Score: {found_student.science_score}")
else:
    print(f"Student with Roll Number {search_roll_number} not found.")

# Display details of students who scored above 90 in Math and Science
students_list.display_high_scorers()
```

```python
#Consider two liked list A and B in sorted order. Write a program to merge the linked list
# A and B such that the final list C is in sorted order.

class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def merge_sorted_lists(list_a, list_b):
    # Create a dummy node to serve as the starting point of the merged list
    dummy_node = Node()
    current = dummy_node

    # Pointers for list A and list B
    current_a, current_b = list_a, list_b

    # Iterate until either list A or list B becomes empty
    while current_a is not None and current_b is not None:
        if current_a.data <= current_b.data:
            current.next = current_a
            current_a = current_a.next
        else:
            current.next = current_b
            current_b = current_b.next

        current = current.next

    # If any elements are remaining in list A, append them to the merged list
    if current_a is not None:
        current.next = current_a

    # If any elements are remaining in list B, append them to the merged list
    if current_b is not None:
        current.next = current_b

    # Return the merged list (starting from the node after the dummy node)
    return dummy_node.next

def print_linked_list(head):
    current = head
    while current:
        print(current.data, end=" -> ")
        current = current.next
    print("None")

# Example usage
list_a = Node(1)
list_a.next = Node(3)
list_a.next.next = Node(5)

list_b = Node(2)
list_b.next = Node(4)
list_b.next.next = Node(6)

print("List A:")
print_linked_list(list_a)

print("\nList B:")
print_linked_list(list_b)

merged_list = merge_sorted_lists(list_a, list_b)

print("\nMerged List C:")
print_linked_list(merged_list)
```
